### What are the years in which home sales took place in 2014 and 2015?

```python
# Count the number of sales for each year and create a bar chart
sales['date'].dt.year.value_counts().plot(kind='bar')

# Set the chart title
plt.title('Number of Sales by Year')

# Set the x-axis label
plt.xlabel('Year')

# Set the y-axis label
plt.ylabel('Number of Sales')
```

### What are all the homes sold in 2014 sorted from earliest to latest?

```python
# Get the count of sales for each month and return the top result
sales[sales.date.dt.year == 2014].sort_values('date', ascending= True)
```

### What is the number of sales that took place in each month of the year?

```python
sales.date.dt.month.value_counts().head(1)
```

### Which homes were sold between May 1st, 2014 and May 1st, 2015?

```python
# Select one year of sales data between May 1, 2014 and May 1, 2015
one_year = sales[sales.date.between('2014-05-01', '2015-05-01')].sort_values('date')

# Count sales for each month and plot as a bar chart
one_year.date.dt.month.value_counts().sort_index().plot(kind='bar')

# Add labels
plt.title('Monthly Sales from May 1, 2014 to May 1, 2015')
plt.xlabel('Month')
plt.ylabel('Number of Sales')
```

### Creating a line plot that shows the total number of sales by week of the year

```python
# Count number of sales by week and generate line plot
weekly_sales = sales.date.dt.isocalendar().week.value_counts().sort_index()
weekly_sales.plot(kind='line')

# Add labels and title to plot
plt.xlabel('Week')
plt.ylabel('Number of Sales')
plt.title('Weekly Sales')

# Display plot
plt.show()
```

### Bar plot showing the total number of sales that took place in December.

```python
# Extract sales data for December, January, and February
df = sales[sales.date.dt.month.isin([12, 1, 2])]

# Count number of sales by month and generate bar plot
monthly_sales = df.date.dt.month.value_counts().sort_index()
monthly_sales.plot(kind='bar')

# Add labels and title to plot
plt.xlabel('Month')
plt.ylabel('Number of Sales')
plt.title('Sales in Winter Months')

# Display plot
plt.show()
```

### What is the correlation between house price and various features?

```python
# Create scatter plot of house price vs. square footage of living area
sns.scatterplot(x='sqft_living', y='price', data=sales)

# Create heatmap of correlation matrix
corr = sales.corr()
sns.heatmap(corr, cmap='coolwarm')
```

### What are the most common zip codes in the dataset?

```python
# Count frequency of each zip code and plot as a bar chart
zip_counts = sales.zipcode.value_counts().sort_index()
zip_counts.plot(kind='bar', figsize=(12,6))
plt.xlabel('Zip Code')
plt.ylabel('Frequency')
plt.show()
```

### What is the average price of houses in different zip codes?

```python
# Group data by zip code and calculate mean house price for each group
zip_price = sales.groupby('zipcode').mean()['price'].sort_values()
print(zip_price)

# Plot mean house price for each zip code
plt.figure(figsize=(12,6))
plt.bar(zip_price.index, zip_price.values)
plt.xticks(rotation=90)
plt.xlabel('Zip Code')
plt.ylabel('Mean House Price')
plt.show()
```
